```python
import pandas as pd
import seaborn as sns
sns.set_context("talk")
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('CIRS_SWS_v_T.csv')
data = data.drop(columns='Temp_F')
data
```

```python
datamelt = data.melt(id_vars=['Temp_C', 'Phantom'], var_name='SWS')
datamelt
```

```python
g = sns.pointplot(x="Temp_C", y="value", hue="Phantom", data=datamelt, markers=['x', 'o', '*', '^'], scale=0.75)
g.set_xlabel('Temperature (C)')
g.set_ylabel('SWS (m/s)')
#plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0., title="Phantom")
#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., title='Phantom')
g.legend(bbox_to_anchor=(0.5, 0.25), loc='center', ncol=2, title='Phantom')
plt.tight_layout()
g.figure.savefig('temp_dependence.png')
```
