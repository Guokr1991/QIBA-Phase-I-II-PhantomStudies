```python
import pandas as pd
import seaborn as sns
sns.set_context("talk")
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('QIBAphase2_longitudinal_VdispVvel_4dates.csv')
data
```

```python
g = sns.FacetGrid(data, col='Phantom', hue='Datatype', size=5, aspect=1)
g.map(plt.errorbar, "Date", "SWSmean", "SWSstd", marker="o")
(g.set_titles("{col_name}").set_axis_labels('', 'SWS (m/s)').set_xticklabels(rotation=30).add_legend())
g.savefig('temporal_stability.png')
```
